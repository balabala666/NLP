# NLTK
NLTK是Python很强大的第三方库，可以很方便的完成很多自然语言处理（NLP）的任务，包括分词、词性标注、命名实体识别（NER）及句法分析

```python
# print('nlp2 test')
import nltk
text = 'PathonTip.com is a very good website. We can learn a lot from it.'
#将文本拆分成句子列表  nltk.sent_tokenize(text)
sens = nltk.sent_tokenize(text)
print(sens)
#将句子进行分词,nltk的分词是句子级的,因此要先分句,再逐句分词,否则效果会很差.
#nltk.word_tokenize(sent) 对句子进行分词
words = []
for sent in sens:
    words.append(nltk.word_tokenize(sent))
print(words)
# tags = []
#词性标注要利用上一步分词的结果
# for tokens in words:
#     tags.append(nltk.pos_tag(tokens))
# print('词性标注：'.tags)
```

```python
import urllib.request
import html5lib
from bs4 import BeautifulSoup
# BeautifulSoup 库来对抓取的文本进行处理
response = urllib.request.urlopen('http://php.net/')
html = response.read()
# print (html)
#利用lxml将抓取的网页转换为干净的文本
soup = BeautifulSoup(html,"lxml")
text = soup.get_text(strip=True)
# print (text)
#文本分词
tokens = [t for t in text.split()]
# print(tokens)
#词频统计
freq = nltk.FreqDist(tokens)
for key,val in freq.items():
    print (str(key) + ':' + str(val))
freq.plot(20,cumulative=False)
# freq.show()
```

```python
import nltk
import urllib.request
import html5lib
from bs4 import BeautifulSoup
# BeautifulSoup 库来对抓取的文本进行处理
from nltk.corpus import stopwords
#过滤一些停止词
response = urllib.request.urlopen('http://php.net/')
html = response.read()
# print (html)
#利用lxml将抓取的网页转换为干净的文本
soup = BeautifulSoup(html,"lxml")
text = soup.get_text(strip=True)
# print (text)
#文本分词
tokens = [t for t in text.split()]
#知识普及：
# L1=L L2=L[:]
# 所以L1=L是把L所指的对象绑定到名字L1上，而L2=L[:]则是把L通过切片运算取得的新列表对象绑定到L2上。
# 前者两个名字指向同一个对象，后者两个名字指向不同对象。
# 换句话说，L1和L是指的同一个东西，那么修改L1也就修改了L；L2则是不同的东西，修改L2不会改变L。
clean_tokens = tokens[:]
sr = stopwords.words('english')
for token in tokens:
    if token in sr:
        clean_tokens.remove(token)
#词频统计
freq = nltk.FreqDist(clean_tokens)
# for key,val in freq.items():
#     print (str(key) + ':' + str(val))
# print(tokens)
freq.plot(20,cumulative=False)
# freq.show()
```
